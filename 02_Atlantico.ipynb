```python
#1.	Faça a leitura de uma imagem 2D colorida qualquer e apresente as informações listadas abaixo por meio da função print().
# altura, largura, Canais de cor, tipo de dado, tom de cinza máximo, tom de cinza médio, tom de cinza mínimo.

import cv2
import numpy as np

# Caminho do arquivo JPG
from google.colab import drive
drive.mount('/content/drive')
local_arquivo ='/content/drive/MyDrive/Atlantico/Bases/frankstein_capa_livro.jpg'

def info_imagem(local_arquivo):

    # Carregar a imagem usando o OpenCV
    imagem = cv2.imread(local_arquivo)

    # Obter informações da imagem
    altura, largura, canais = imagem.shape
    tipo_dado = imagem.dtype
    tom_maximo = np.max(imagem)
    tom_medio = np.mean(imagem)
    tom_minimo = np.min(imagem)

    # Imprimir as informações
    print(f"Altura: {altura}")
    print(f"Largura: {largura}")
    print(f"Canais de cor: {canais}")
    print(f"Tipo de dado: {tipo_dado}")
    print(f"Tom de cinza máximo: {tom_maximo}")
    print(f"Tom de cinza médio: {tom_medio}")
    print(f"Tom de cinza mínimo: {tom_minimo}")

#Chamar a função
info_imagem(local_arquivo)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Caminho do arquivo JPG
#from google.colab import drive
#drive.mount('/content/drive')
local_arquivo ='/content/drive/MyDrive/Atlantico/Bases/frankstein_capa_livro.jpg'


def aplicar_filtros(local_arquivo):
    # Carregar a imagem usando o OpenCV
    imagem = cv2.imread(local_arquivo)

    # Converter a imagem para escala de cinza
    imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

    # Aplicar filtro da média
    imagem_media = cv2.blur(imagem_cinza, (5, 5))

    # Aplicar filtro da mediana
    imagem_mediana = cv2.medianBlur(imagem_cinza, 5)

    # Aplicar filtro gaussiano
    imagem_gaussiano = cv2.GaussianBlur(imagem_cinza, (5, 5), 0)

    # Exibir as imagens usando Matplotlib
    plt.figure(figsize=(10, 10))

    plt.subplot(2, 2, 1)
    plt.imshow(imagem_cinza, cmap='gray')
    plt.title('Imagem Original')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.imshow(imagem_media, cmap='gray')
    plt.title('Filtro da Média')
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.imshow(imagem_mediana, cmap='gray')
    plt.title('Filtro da Mediana')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.imshow(imagem_gaussiano, cmap='gray')
    plt.title('Filtro Gaussiano')
    plt.axis('off')

    plt.show()

#Chamar a função
aplicar_filtros(local_arquivo)
```
